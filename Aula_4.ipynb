```python
import pandas as pd
```

```python
def process_sdg_data(csv_file, columns_to_drop):
  df = pd.read_csv(csv_file)
  df = df.drop(columns_to_drop, axis=1)
  df = df.set_index("Country Name").transpose()
  return df
```
